```python
df.loc[(df.valence <=0.5) &(df.danceability<=0.5) ,'control'] = 'low low'
df.loc[(0.5<df.valence) & (df.danceability<=0.5),'control'] = 'high low'
df.loc[(df.valence<=0.5)&(0.5<df.danceability),'control']='low high'
df.loc[(0.5<df.valence)&(0.5<df.danceability),'control']='high high'
df.loc[df.energy <=0.25 ,'Energy'] = 'low'
df.loc[(0.25<df.energy) & (df.energy<=0.5),'Energy'] = 'slightly low'
df.loc[(df.energy<=0.75)&(0.5<df.energy),'Energy']='slightly high'
df.loc[0.75<df.energy,'Energy']='high'
df.loc[df.acousticness <=0.25 ,'Acousticness'] = 'electric'
df.loc[(0.25<df.acousticness) & (df.acousticness<=0.5),'Acousticness'] = 'slightly electric'
df.loc[(df.acousticness<=0.75)&(0.5<df.acousticness),'Acousticness']='slightly acoustic'
df.loc[0.75<df.acousticness,'Acousticness']='acoustic'
```
